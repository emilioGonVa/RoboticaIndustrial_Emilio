```python
a = 1
b = 0
x = np.arange(-10,10,1)
y = a*x+b
plt.plot(x,y)
```

```python
x = np.arange(0,10,0.001)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.arange(0,10,0.001)
y = np.exp(x)
plt.plot(x,y,'yellow')
```

```python
x = np.arange(0,10,0.001)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,'yellow')
plt.plot(x,y2,'--g',linewidth=2)
```

```python
x = np.arange(-10,20,0.001)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,'yellow')
plt.plot(x,y2,'--g',linewidth=2)
plt.title("Prueba de grficas en python")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axis([-10,20,-1,1])
```

```python
x = np.arange(-10,20,0.001)
y = np.sin(x)
y2 = np.cos(x)
y3 = 5*x
y4 = np.exp(x)
plt.subplot(2,2,1)
plt.plot(x,y,label = 'line1')
plt.legend(loc=0)
plt.subplot(2,2,2)
plt.plot(x,y2,label = 'line2')
plt.legend(loc=0)
plt.subplot(2,2,3)
plt.plot(x,y3,label = 'line3')
plt.legend(loc=0)
plt.subplot(2,2,4)
plt.plot(x,y4, label = 'line4')
plt.legend(loc=0)
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib nbagg
%matplotlib nbagg
fig = plt.figure()
ax = fig.gca(projection = '3d')

x = np.arange(-10,10,0.1)
y = np.sin(x)
z = np.cos(x)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Grfica 3D')
ax.plot(x,y,z,c='red')
ax.view_init(20,70)
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)
xv,yv = np.meshgrid(x,y)
zv = xv**2+yv**2

ax.plot_wireframe(xv,yv,zv)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Grfica 3D')
ax.view_init(10,10)
```
